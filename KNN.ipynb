```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
```

```python
X , y =make_blobs(n_samples=2000 , n_features=2 , cluster_std=3 , centers=3 , random_state=42)
n_features = X.shape[1]
m = X.shape[0]
```

```python

xt = np.array([0.8 ,-3])
for i in range(m):
  if y[i]==0:
    plt.scatter(X[i][0] , X[i][1] , color='red')
  elif y[i]==1:
    plt.scatter(X[i][0] , X[i][1] , color='green')
  else:
    plt.scatter(X [i][0] , X[i][1] , color='blue')

plt.scatter(xt[0] ,xt[1] , color='orange', marker="*")
plt.show()
```

```python
def dist (p,q):
  return np.sqrt(np.sum((p-q)**2))

def knn(X, y , xt , k=7):

  m=X.shape[0]
  dist_arr = []

  for i in range(m):
    d=dist(X[i] , xt)
    dist_arr.append((d , y[i]))

  sortred_dist = sorted(dist_arr)
  sortred_dist = np.array(sortred_dist[:k])
  labels=sortred_dist[: , 1]

  labels, cnts=np.unique(labels , return_counts=True)
  idx=cnts.argmax()
  pred=labels[idx]
  return int(pred)


```

```python
knn(X, y , xt )
```
